```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from yahooquery import Ticker

# Lista de ativos fixos
ATIVOS = ["EURGBP=X", "EURCAD=X", "USDCAD=X"]
PERIODO = "23mo"
INTERVALO = "1d"

# Função para obter dados históricos
def obter_dados_ativo(par_moeda):
    ativo = Ticker(par_moeda)
    dados_historicos = ativo.history(period=PERIODO, interval=INTERVALO)

    if isinstance(dados_historicos, pd.DataFrame):
        dados_historicos.reset_index(inplace=True)
        return dados_historicos
    else:
        print(f"Erro ao obter os dados do ativo {par_moeda}.")
        return None

# Função para calcular indicadores técnicos
def calcular_indicadores(dados):
    dados['Média Móvel 20'] = dados['close'].rolling(window=20).mean()
    dados['RSI'] = calcular_rsi(dados['close'])
    return dados

# Função para calcular RSI
def calcular_rsi(precos, periodo=14):
    delta = precos.diff()
    ganho = np.where(delta > 0, delta, 0)
    perda = np.where(delta < 0, abs(delta), 0)

    ganho_medio = pd.Series(ganho).rolling(periodo).mean()
    perda_media = pd.Series(perda).rolling(periodo).mean()

    rs = ganho_medio / perda_media
    rsi = 100 - (100 / (1 + rs))
    return pd.Series(rsi)

# Função para visualizar dados
def plotar_dados(dados, par_moeda):
    plt.figure(figsize=(12, 6))

    # Gráfico de preço e médias móveis
    plt.subplot(2, 1, 1)
    plt.plot(dados.index, dados['close'], label="Preço de Fechamento", color="blue")
    plt.plot(dados.index, dados['Média Móvel 20'], label="Média Móvel 20", color="red")
    plt.legend()
    plt.title(f"Análise do Ativo {par_moeda}")

    # Gráfico RSI
    plt.subplot(2, 1, 2)
    plt.plot(dados.index, dados['RSI'], label="RSI", color="purple")
    plt.axhline(y=70, color='red', linestyle='--')  # Nível de sobrecompra
    plt.axhline(y=30, color='green', linestyle='--')  # Nível de sobrevenda

    plt.legend()
    plt.xlabel("Tempo")

    plt.tight_layout()
    plt.show()

# Executando análise para cada ativo
for ativo in ATIVOS:
    dados = obter_dados_ativo(ativo)

    if dados is not None:
        dados = calcular_indicadores(dados)
        plotar_dados(dados, ativo)
```
